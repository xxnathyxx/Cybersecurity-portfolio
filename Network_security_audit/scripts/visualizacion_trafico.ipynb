# 📊 Análisis de Tráfico de Red
Este notebook muestra visualizaciones de ejemplo basadas en el tráfico capturado.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar muestra de tráfico
df = pd.read_csv('../lab/traffic_sample.csv')
df.head()
```

```python
# Conteo por protocolo
protocol_counts = df['protocol'].value_counts()
protocol_counts.plot(kind='bar', figsize=(6,4), title='Distribución por Protocolo')
plt.show()
```

```python
# Top 5 IP origen
top_ips = df['src_ip'].value_counts().head(5)
top_ips.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6), title='Top 5 IP Origen')
plt.ylabel('')
plt.show()
```
